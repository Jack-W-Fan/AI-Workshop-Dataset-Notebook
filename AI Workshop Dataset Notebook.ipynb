# **AI Workshop Dataset Creation**
In this workshop, we are just going to quickly go through a couple lines of simple code you can use to quickly generate a simple, but visually appealling, graph for your research.

Our first step is to upload our dataset, which we have provided.

First, just because we are doing this locally, we have to install pandas, which is one of the most popular Python libraries used in most relevant coding areas.

```python
pip install pandas
```

Once this is installed (you can do this locally, later on) we have to import our necessary commands to run them.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Next, we will load the dataset into our code, so it is usable.

```python
df = pd.read_csv("data-table.csv")
```

Here, we can run a simple command to show us what the csv we are uploading looks like.

```python
df.head(50)
```

Next, we will write a line that will convert our dates column into something that is readable by the program.

```python
df["week_end"] = pd.to_datetime(df["week_end"])
```

Now, since we are focusing on COVID-19 here, and we have multiple other pathogens, we want to filter for only COVID-19. Here, we have set "covid" as our variable, which we will be able to modify/use later on in our code.

```python
covid = df[df["pathogen"] == "COVID-19"]
```

Now, with our data fully readable and our columns set, we can make the graph. Here, we are going to use the plt function, which we imported earlier.

```python
# Create a new figure for the plot
plt.figure()

# Plot children's COVID-related visits over time
plt.plot(covid["week_end"], covid["Children"])

# Label the x-axis (time)
plt.xlabel("Week Ending")

# Label the y-axis (percentage of visits in children)
plt.ylabel("Percent of Children Vaccinated")

# Add a title to the plot
plt.title("COVID-19 Vaccinations of Children Over Time")

# Changes the date to a readable format
plt.gcf().autofmt_xdate()

# This will save our graph locally so we can download and use our graph
plt.savefig("covid_children_over_time.png")

# Display the plot
plt.show()
```

Similarly, instead of a line graph, if we wanted to make a bar graph instead, we would only have to change a couple lines of code!

```python
plt.figure()

# Here, instead of plt.plot which we used before, we will use plt.bar. We will also add the ", width=6" to set the width of each bar
plt.bar(covid["week_end"], covid["Children"], width=6)
plt.xlabel("Week Ending")
plt.ylabel("Percent of Children Vaccinated")
plt.title("COVID-19 Vaccinations of Children Over Time")
plt.gcf().autofmt_xdate()

# Here, we can just change the name of the png generated so that they are separate.
plt.savefig("covid_children_over_time_bar.png")
plt.show()
```
