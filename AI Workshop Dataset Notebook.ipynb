```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load the data
df = pd.read_csv("data-table.csv")

# 2. Convert date column
df["week_end"] = pd.to_datetime(df["week_end"])

# 3. Filter to one pathogen
covid = df[df["pathogen"] == "COVID-19"]

# 4. Make the plot
plt.figure()
plt.plot(covid["week_end"], covid["Children"])
plt.xlabel("Week Ending")
plt.ylabel("Percent of Visits (Children)")
plt.title("COVID-19 Related Visits in Children Over Time")

# 5. Show the graph
plt.show()
```
